### 1)Загружаем необходимые библиотеки

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

###2) Подготовка данных:

* Загрузка данных из моего репозитория GitHub.
* Переименовываем столбец 'Unnamed: 0' в 'date'
* Приводем данные столбца 'date' к типу дата и делаем его индексом

```python
data = pd.read_csv('https://raw.githubusercontent.com/gainutdinov-rafael/architectAI/main/certification1/dataset_tk.csv')
data = data.rename(columns={'Unnamed: 0':'date'})
data['date'] = pd.to_datetime(data['date'], format='%d/%m/%Y %H:%M:%S')
data.set_index(['date'],inplace=True)
data.head(1000)
```

###3) Дополнение таблица необходимыми данными

* добавим и расчитаем столбец общего потребления электроэнергии по всей стране
* добавим и расчитаем среднее потребление электтроэнергии по всей стране

```python
data = data.assign(sum=0)
data = data.assign(mean=0)
for col in data.columns:
  data['sum']=data['sum']+data[col]
data['mean'] = data['sum']/len(data.columns)
```

###4) Ищем выбросы

В статистике выбросы — это значения, резко отличающиеся от других значений в собранном наборе данных. Выброс может указывать на аномалии в распределении данных или на ошибки при измерениях, поэтому зачастую выбросы исключаются из набора данных.

###Графики для поиска выбросов по группам

Группы по потреблению: первый график - 0-10 гВт, второй - 10-50 гВт, третий - 50-500 гВт

```python
sns.boxplot(data[['Punjab','Haryana','Rajasthan','Delhi','UP','Gujarat','MP','Maharashtra','Andhra Pradesh','Telangana','Karnataka','Tamil Nadu','Bihar','Odisha','West Bengal']],width=0.8,orient='h')
```

###5) Из графиков видно большое колиество выбросов в некоторых штатах. Для их отчистки применим 2 способа:
* Очистка методом: z-оценка
* Очистка методом межквартильного диапазона

```python
z = np.abs(stats.zscore(data))
data_clean1 = data[(z<3).all(axis=1)]
data_clean1.shape
```

```python
Q1 = data.quantile(q=.25)
Q3 = data.quantile(q=.75)
IQR = data.apply(stats.iqr)
data_clean2 = data[~((data < (Q1-1.5*IQR)) | (data > (Q3+1.5*IQR))).any(axis=1)]
data_clean2.shape
```

В первоначальном наборе даннных было 503 записи. При очистке по z-оценке осталось 440 записей, а при очистке методом межквартильного диапазона осталось 249 записей.

Ни один из способов не является 100% верным, т.к. нельзя однозначно сказать является то или иное значение ошибочным, но исключая выбросы можно получить новую информацию.

###График суммарного потребления электроэнергии штатами в Индии

график суммарного потребления нужен чтобы видеть "картину экономики в целом".

```python
fig,axes = plt.subplots(nrows=1)
fig.suptitle('График суммарного потребления энергии',fontsize=22)
data['sum'].plot(x=data.index,y=data['sum'],figsize=(30,10),label='До очистки',ax=axes,lw=5,linestyle='-',marker='o',markerfacecolor='black')
data_clean1['sum'].plot(x=data_clean1.index,y=data_clean1['sum'],figsize=(30,10),label='Очистка z-оценкой',ax=axes,lw=5,linestyle=':',marker='v',markerfacecolor='red')
data_clean2['sum'].plot(x=data_clean2.index,y=data_clean2['sum'],figsize=(30,10),label='Очистка методом межквартильного диапазона',ax=axes,lw=5,linestyle=':',marker='o',markerfacecolor='blue')

axes.legend(fontsize=12)
axes.grid(linestyle='--', alpha=0.8,axis='y')

#data['sum'].plot(x=data.index,figsize=(30,10))
```

Глядя на график можно сделать вывод, что что то произошло в июне 2020 года, после чего график стал "ступенчатым". Если вспомнить события 2020 года, то можно предположить, что причиной такого изменения стала пандемия Ковид2019.

На следующем графике сравним потребление двумя случайными штатами и сравним его со средними знаениями.

```python
fig,axes = plt.subplots(nrows=1)
data['UP'].plot(x=data.index,figsize=(30,10),label='UP')
data['Maharashtra'].plot(x=data.index,figsize=(30,10),label='Maharashtra')
data['mean'].plot(x=data.index,figsize=(30,10),label='Среднее значение')
plt.legend(fontsize=12)
plt.show()
```

По графику можно пронаблюдать периодическое расхождение в "разные стороны", т.е. когда в одном штате потребление растет, в другом потребление электроэнергии падает. Махараштра — лидирующий промышленный штат страны, Уттар-Прадеш — один из наиболее отсталых в экономическом отношении штатов страны. В то же время, Уттар Прадеш — вторая в стране экономика (после Махараштры).

В период пандемии Ковид19 в сентябре-октябре 2020 года. Махараштра резко увеличила потребление электоэнергии, что может говорить о запуске промышленности, а Уттар-Прадеш наоборот сильно упало потребление, что является признаком падения производства как основного потребителя электроэнергии. Т.е. Махараштра является более экономически устойчивым штатом чем Уттар-Прадеш.

Добавим график зависомости потребления электроэнергии штта от среднего значения потребления электроэнергии по всем штатам, для того чтобы посмотреть на аномалии.

```python
fig,axes = plt.subplots()
#axes.scatter(data['mean'], data['Pondy'],label='Pondy')
axes.scatter(data['mean'], data['Sikkim'],label='Sikkim')
axes.scatter(data['mean'], data['Tripura'],label='Tripura')
axes.scatter(data['mean'], data['Chandigarh'],label='Chandigarh')
plt.legend(fontsize=12)
plt.show()
```

по графику видно что иногда встречаются аномалии, но в целом разброс не значительный.

Попробуем посмотреть на какие периоды выпадало наибольшее потребление электроэнергии. Оценку будем проводить по среднему значению.

Графики потребления по всем штатам. 35 графиков в одном размещать не удобно, поэтому сделал разделение на несколько графиков. Через переменную **ax** можно установить необходимое количество графиков для разбивки.

```python
ax = 3 #Количество графиков
fig,axes = plt.subplots(nrows=ax)
fig.suptitle('График потребления энергии по штатам',fontsize=22)
fig.set_dpi(1000)
i=0
for col in data:
  if col != 'sum':
    data[col].plot(x=data.index,y=data[col],figsize=(30,30),label=col,ax=axes[i%ax])
    i += 1

for i in range(0,ax):
  axes[i].legend(fontsize=12)
  axes[i].grid(linestyle='--', alpha=0.8,axis='y')
```
